```python
#1. 패션 mnist 다운로드
from tensorflow import keras

(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()

print(train_input.shape, train_target.shape)
print(test_input.shape, test_target.shape)
```

```python
#2. 데이터 정규화(0~1)
train_scaled = train_input / 255.0
test_scaled = test_input / 255.0

print(train_scaled.shape, test_scaled.shape)
```

```python
#훈련데이터 20%를 검증데이터 설정, shape 출력
from sklearn.model_selection import train_test_split

train_scaled, val_scaled, train_target, val_target = train_test_split(train_scaled, train_target,test_size=0.2, random_state=42)

print(train_scaled.shape, test_scaled.shape, val_scaled.shape)
```

```python
#4. Flatten 사용 네트워크 생성
from tensorflow import keras

model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape=(28,28), name='flatten'))
model.add(keras.layers.Dense(100, activation='relu', name='hidden1'))
model.add(keras.layers.Dense(64, activation='relu', name='hidden2'))
model.add(keras.layers.Dense(32, activation='relu', name='hidden3'))
model.add(keras.layers.Dense(10, activation='softmax', name='output'))

model.summary()
```

```python
#5. 모델 훈련
model.compile(optimizer='sgd', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_scaled, train_target, epochs=50)
```

```python
#6. 검증/테스트 데이터 점수 확인
model.evaluate(val_scaled, val_target)
model.evaluate(test_scaled, test_target)
```

```python
#7. 옵티마이저를 adam으로 변경, 훈련 후 점수 확인

new_model = keras.Sequential()
flatten = keras.layers.Flatten(input_shape=(28,28), name='flatten')
hidden1 = keras.layers.Dense(100, activation='relu', name='hidden1')
hidden2 = keras.layers.Dense(64, activation='relu', name='hidden2')
hidden3 = keras.layers.Dense(32, activation='relu', name='hidden3')
output = keras.layers.Dense(10, activation='softmax', name='output')

new_model.add(flatten)
new_model.add(hidden1)
new_model.add(hidden2)
new_model.add(hidden3)
new_model.add(output)

new_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
new_model.fit(train_scaled, train_target, epochs=50)

new_model.evaluate(val_scaled, val_target)
new_model.evaluate(test_scaled, test_target)
```
