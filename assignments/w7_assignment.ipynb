```python
#1번 문제
import pandas as pd

#생선 데이터를 읽고, fish_data에 저장
fish_data = pd.read_csv('https://bit.ly/fish_csv_data')

#특성 선택(Weight, Length, Height, Width) + 첫 100개 샘플만 사용
features = ['Weight', 'Length', 'Height', 'Width']
fish_input = fish_data[features].iloc[:100].to_numpy()
```

```python
#2번 문제
#타깃을 fish_target에 저장
fish_target = fish_data['Species'].iloc[:100].to_numpy()
```

```python
#3번 문제
from sklearn.model_selection import train_test_split

#훈련데이터와 테스트데이터로 나누고 shape을 출력
train_input, test_input, train_target, test_target = train_test_split(fish_input, fish_target, random_state=42)
print(train_input.shape, test_input.shape)
print(train_target.shape, test_target.shape)
```

```python
#4번 문제
#훈련데이터와 테스트데이터를 표준화
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
#5번 문제
#로지스틱 회귀모델을 생성, 훈련 후 훈련데이터와 테스트데이터의 점수를 출력
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)
lr.fit(train_scaled, train_target)
print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))
```

```python
#6번 문제
import numpy as np

#테스트데이터의 첫 3개의 샘플에 대해 분류 결과와 각 클래스의 확률 출력
print(lr.predict(test_scaled[:3]))
proba = lr.predict_proba(test_scaled[:3])
print(np.round(proba, decimals=3))
```

```python
#7번 문제
from scipy.special import softmax

#테스트데이터의 첫 3개의 샘플에 대해 선형방정식의 결과 z 출력
decision = lr.decision_function(test_scaled[:3])
print('z:\n', np.round(decision, 3))

#소프트맥스 함수 이용 확률 출력
proba = softmax(decision, axis=1)
print(np.round(proba, decimals=3))
```
