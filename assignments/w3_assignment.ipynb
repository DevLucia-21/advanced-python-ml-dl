```python
#10개의 1로 이루어진 1차원 배열 arr1 생성
import numpy as np
arr1 = np.ones(10)
print(arr1)
```

```python
#10개의 0으로 이루어진 1차원 배열 arr2 생성
arr2 = np.zeros(10)
print(arr2)
```

```python
#arr1과 arr2를 연결하여 arr3를 생성
arr3 = np.concatenate((arr1, arr2))
print(arr3)
```

```python
#3으로 채워진 3행 5열인 배열 arr4를 생성
arr4 = np.full((3, 5), 3)
print(arr4)
```

```python
#arr1과 arr2를 열방향으로 연결한 배열 arr5 생성
arr5 = np.column_stack((arr1, arr2))
print(arr5)
```

```python
#arr1과 arr2를 행방향으로 연결한 배열 arr6 생성
arr6 = np.row_stack((arr1, arr2))
print(arr6)
```

```python
#sklearn의 적절한 함수를 이용하여 아래의 데이터를 훈련데이터와 테스트 데이터로 나누기
#data는 운동선수의 키와 몸무게로 이루어진 배열임(numpy 이용할 것)
#유도선수는 1, 경마선수는 2(? 0인듯)
#data = [[180,80],[175,70],[190,88],[190,80],[187,92],[160,54],[165,52],[155,49],[153,43]]
#target = [1, 1, 1, 1, 1, 0, 0, 0, 0]

from sklearn.model_selection import train_test_split
data = np.array([[180, 80], [175, 70], [190, 88], [190, 80], [187, 92], [160, 54], [165, 52], [155, 49], [153, 43]])
target = np.array([1, 1, 1, 1, 1, 0, 0, 0, 0])
train_input, test_input, train_target, test_target = train_test_split(data, target, stratify=target, random_state=42)
print(train_input.shape, test_input.shape)
print(train_target.shape, test_target.shape)
print(test_target)
```

```python
#data를 산점도로 표현
import matplotlib.pyplot as plt
plt.scatter(data[:,0], data[:,1])
plt.xlabel('height')
plt.ylabel('weight')
plt.show()
```

```python
#데이터를 표준점수로 변환해서 scaled_data에 저장
mean = np.mean(train_input, axis=0)
std = np.std(train_input, axis=0)
print(mean, std)
scaled_data = (train_input - mean) / std
print(scaled_data)
```

```python
#scaled_data를 산점도로 표현
plt.scatter(scaled_data[:,0], scaled_data[:,1])
plt.xlabel('height')
plt.ylabel('weight')
plt.show()
```
