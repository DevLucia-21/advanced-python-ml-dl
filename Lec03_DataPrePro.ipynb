```python
#넘파이에서 어레이를 생성하고, 연결하는 다양한 함수
import numpy as np

arr1 = np.ones(5)
arr2 = np.full(5,2)
print(arr1,'\n', arr2)
print(np.column_stack((arr1, arr2))) #column_stack은 인자 1개라 ()로 묶어줘야함
print(np.row_stack((arr1, arr2)))
np.concatenate((arr1, arr2))
print("****************************")

arr3 = np.ones((2,3))
arr4 = np.full((2,3),2)
print(arr3,'\n', arr4)
print(np.column_stack((arr3, arr4)))
print("----------------------------")
print(np.row_stack((arr3, arr4)))
print("----------------------------")
print(np.concatenate((arr3, arr4), axis=0)) #0은 행방향
print("----------------------------")
print(np.concatenate((arr3, arr4), axis=1)) #1은 열방향
```

```python
#도미의 길이와 무게
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]

#빙어의 길이와 무게
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

#도미와 빙어 데이터 병합
fish_length = bream_length + smelt_length
fish_weight = bream_weight + smelt_weight
print(fish_length[:5])
print(fish_weight[:5])
```

```python
fish_data = np.column_stack((fish_length, fish_weight))
fish_target = np.concatenate((np.ones(35), np.zeros(14)))
print(fish_data[:5])
print(fish_target)
```

```python
#train_test_split를 이용하여 데이터를 훈련데이터와 테스트데이터로 나눔
from sklearn.model_selection import train_test_split
#train_input, test_input, train_target, test_target = train_test_split(fish_data, fish_target, random_state=42)
train_input, test_input, train_target, test_target = train_test_split(fish_data, fish_target, stratify=fish_target, random_state=42)
print(train_input.shape, test_input.shape)
print(train_target.shape, test_target.shape)
print(train_target)
print(test_target)
print(np.count_nonzero(fish_target)/np.count_nonzero(1-fish_target))
print(np.count_nonzero(train_target)/np.count_nonzero(1-train_target))
print(np.count_nonzero(test_target)/np.count_nonzero(1-test_target))
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(train_input, train_target)
kn.score(test_input, test_target)
new_input = [[25,150]]
kn.predict(new_input)
#new_input의 5개 이웃 정보(거리, 이웃번호)
dist,idx = kn.kneighbors(new_input)
print(dist,idx)
print(train_input[idx])
```

```python
import matplotlib.pyplot as plt
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(new_input[0][0], new_input[0][1], marker='^')
plt.scatter(train_input[idx,0], train_input[idx,1], marker='D')
plt.xlabel('length')
plt.ylabel('weight')
plt.xlim(0,1000)
plt.ylim(0,1000)
plt.show()
```

```python
#두 특성(길이, 무게)의 스케일을 맞추기 위해서 표준점수(z)를 사용
#z는 특성값이 평균에서 표준편차의 몇 배인지를 나타냄
mean = np.mean(train_input, axis=0)
std = np.std(train_input, axis=0)
print(mean, std)
train_scaled = (train_input-mean)/std
test_scaled = (test_input-mean)/std
```

```python
kn.fit(train_scaled, train_target)
print(kn.score(train_scaled, train_target))
print(kn.score(test_scaled, test_target))
```
