```python
from tensorflow import keras
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
df_iris = pd.DataFrame(data = iris.data, columns=iris.feature_names)
df_iris['target'] = iris.target
df_iris.head()
data = df_iris.iloc[:,:-1].to_numpy()
target = df_iris.iloc[:,-1].to_numpy()
print(data[:5],target[:5])
print(data.shape,target.shape)
```

```python
train_input, test_input, train_target, test_target = train_test_split(data, target, test_size=0.2, random_state=42)
print(train_input.shape, test_input.shape)
```

```python
sd = StandardScaler()
sd.fit(train_input)
train_input = sd.transform(train_input)
test_input = sd.transform(test_input)
```

```python
#로지스틱회귀
from sklearn.linear_model import SGDClassifier
sgd = SGDClassifier(loss='log_loss',max_iter=500,tol=1e-3)
sgd.fit(train_input,train_target)
print(sgd.score(train_input,train_target))
print(sgd.score(test_input,test_target))
```

```python
#ANN
dense = keras.layers.Dense(3,activation='softmax',input_shape=(4,))
model = keras.Sequential([dense])
model.compile(loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(train_input,train_target,epochs=500)
```

```python
print(model.evaluate(train_input,train_target))
print(model.evaluate(test_input,test_target))
```

```python
pred = model.predict(test_input)
print(np.round(pred[:5],decimals=3))
pred_idx = np.argmax(pred,axis=1)
print(pred_idx)
print(test_target)
```
