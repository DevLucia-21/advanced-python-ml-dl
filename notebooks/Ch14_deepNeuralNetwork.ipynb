```python
#패션 mnist 다운로드
from tensorflow import keras
(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()
```

```python
print(train_input.shape, train_target.shape)
print(test_input.shape, test_target.shape)
```

```python
#훈련데이터를 훈련데이터 80%, 검증데이터 20%로 나눔
from sklearn.model_selection import train_test_split
train_scaled = train_input / 255.0
train_scaled = train_scaled.reshape(-1, 28*28)
test_scaled = test_input / 255.0
test_scaled = test_scaled.reshape(-1, 28*28)
train_scaled, val_scaled, train_target, val_target = train_test_split(train_scaled, train_target,test_size=0.2, random_state=42)
print(train_scaled.shape, val_scaled.shape)
```

```python
#신경망의 레이어 정의
dense1 = keras.layers.Dense(100, activation='sigmoid', input_shape=(784,)) #입력층 노드(784) - 은닉층 노드(100)
dense2 = keras.layers.Dense(10, activation='softmax') #은닉층 노드(100) - 출력층 노드(10)
dense1.name = 'input'
dense2.name = 'output'
```

```python
#신경망 모델 생성
model = keras.Sequential([dense1, dense2])
```

```python
#모델의 정보 출력
model.summary()
```

```python
#층을 추가하는 다른 방법1
#장점 => 한 눈에 레이어를 알아볼 수 있어 편리하나
#단점 => 조건에 따라 층을 추가하기 힘들고, 많은 층을 추가하려면 생성자가 너무 길어짐
model1 = keras.Sequential([
    keras.layers.Dense(100, activation='sigmoid', input_shape=(784,), name='hidden'),
    keras.layers.Dense(10, activation='softmax', name='output')
], name='fashion MNIST model')
```

```python
model1.summary()
```

```python
#층을 추가하는 다른 방법2: add() 함수를 이용하여 층을 추가함
#=> 한 눈에 레이어 정보를 보기 쉽고, 레이어를 동적으로 추가할 수 있음
model2 = keras.Sequential()
model2.add(keras.layers.Dense(100, activation='sigmoid', input_shape=(784,), name='hidden'))
model2.add(keras.layers.Dense(10, activation='softmax', name='output'))
```

```python
model2.summary()
```

```python
#모델 훈련
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_scaled, train_target, epochs=5)
```

```python
model.evaluate(val_scaled, val_target)
model.evaluate(test_scaled, test_target)
```

```python
#새로운 활성화 함수 relu의 적용: relu(z) = max(0.z)
#입력층의 차원을 1차원으로 펼치는 Flatten을 사용
model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape=(28,28), name='flatten'))
model.add(keras.layers.Dense(100, activation='relu', name='hidden'))
model.add(keras.layers.Dense(10, activation='softmax', name='output'))
```

```python
model.summary()
```

```python
#flatten 층이 모델에 추가되었으므로 데이터는 reshape를 수행하지 않고 준비
(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()
train_scaled = train_input / 255.0
#불필요 => train_scaled = train_scaled.reshape(-1, 28*28)
test_scaled = test_input / 255.0
#불필요 => test_scaled = test_scaled.reshape(-1, 28*28)
train_scaled, val_scaled, train_target, val_target = train_test_split(train_scaled, train_target,test_size=0.2, random_state=42)
print(train_scaled.shape, val_scaled.shape, test_scaled.shape)
```

```python
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_scaled, train_target, epochs=5)
```

```python
val_score = model.evaluate(val_scaled, val_target)
test_score = model.evaluate(test_scaled, test_target)
print(val_score)
print(test_score)
```

```python
#adam optimizer를 사용한 새 모델 정의
#adam: 적응적 학습률 + 모멘텀 최적화를 사용한 옵티마이저
new_model = keras.Sequential()
flatten = keras.layers.Flatten(input_shape=(28,28), name='flatten')
hidden = keras.layers.Dense(100, activation='relu', name='hidden')
output = keras.layers.Dense(10, activation='softmax', name='output')
new_model.add(flatten)
new_model.add(hidden)
new_model.add(output)
```

```python
new_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
new_model.fit(train_scaled, train_target, epochs=5)
```

```python
val_score = new_model.evaluate(val_scaled, val_target)
test_score = new_model.evaluate(test_scaled, test_target)
print(val_score)
print(test_score)
```
